```python
import geopandas as gpd
import contextily as cx
import json
```

```python
with open('secrets/maptiler.json') as fh:
    api_key_maptiler = json.load(fh)['api_key']
```

```python
gdf_1918 = (
    gpd.read_file('data/medway_historic_boundaries/1885_constituencies___england__south_.geojson')
    .loc[lambda _df: _df['Name'].isin(['Chatham', 'Rochester', 'Medway', 'Maidstone', 'Faversham'])]
)
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots(figsize=(20,20))
gdf_1918.boundary.plot(ax=ax, color='green')

cx.add_basemap(ax,crs=gdf_1918.crs, source=f"https://api.maptiler.com/tiles/uk-osgb10k1888/{{z}}/{{x}}/{{y}}.jpg?key={api_key_maptiler}")
gdf_1918.apply(lambda x: ax.annotate(text=x['Name'], xy=x.geometry.centroid.coords[0], ha='center', size=15), axis=1)

plt.title('Medway - Parliamentary Constituencies - 1885', loc='left', size=30)

plt.savefig('Medway - Parliamentary Constituencies - 1885.png', bbox_inches='tight')
```
